```python
# Sample sentences from different domains
sentences = [
    "Virat Kohli scored a century in the match",
    "India won the cricket match comfortably",
    "Cybersecurity protects systems from attacks",
    "Hackers use malware to exploit vulnerabilities",
    "Cooking pasta requires boiling water",
    "Italian food is known for pasta and pizza"
]

# Load model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Generate embeddings
embeddings = model.encode(sentences)

print("Embedding shape:", embeddings.shape)
```

```python
# Compute cosine similarity
similarity_matrix = cosine_similarity(embeddings)

# Display similarity matrix
print("Cosine Similarity Matrix:\n")
print(np.round(similarity_matrix, 2))
```

```python
# Find most similar sentences
for i in range(len(sentences)):
    similarities = similarity_matrix[i]
    most_similar_index = similarities.argsort()[-2]  # second highest (excluding itself)

    print(f"\nSentence: {sentences[i]}")
    print(f"Most similar: {sentences[most_similar_index]}")
```
